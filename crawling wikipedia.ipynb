```python
#imports
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#crawling data from wikipedia with bs4
URL="https://en.wikipedia.org/wiki/List_of_airlines_of_the_United_States"
Page = urllib.request.urlopen(URL)
Soup = BeautifulSoup(Page, "lxml")
Right_table=Soup.find_all('table', class_='wikitable')
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
for row in Right_table[0].findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==8:
        A.append(cells[0].find(text=True))
        B.append(cells[2].find(text=True))
        C.append(cells[3].find(text=True))
        D.append(cells[4].find(text=True))
        E.append(cells[5].find(text=True))
        F.append(cells[6].find(text=True))
        G.append(cells[7].find(text=True))
```

```python
df=pd.DataFrame(A,columns=['Airline'])
df['IATA']=B
df['ICAO']=C
df['Callsign']=D
df['Primary Hubs, Secondary Hubs']=E
df['Founded']=F
df['Notes']=G
df = df.replace('\n','', regex=True)
df.to_csv("Airplane_Data.csv")
df
```
