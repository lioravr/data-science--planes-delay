```python
#start crawling api from kaggle
from kaggle.api.kaggle_api_extended import KaggleApi
from zipfile import ZipFile
Api = KaggleApi()
Api.authenticate()
```

```python
#crawling data for map (Visualisation)
Api.dataset_download_file('usdot/flight-delays','airports.csv')
Airplanes=pd.read_csv('airports.csv')
Airplanes.drop(['AIRPORT','CITY','STATE','COUNTRY'], axis=1, inplace=True)
Airplanes.to_csv("LocationAirports.csv")
```

```python
def GetApi(x):
    Api.dataset_download_file('yuanyuwendymu/airline-delay-and-cancellation-data-2009-2018',str(x)+'.csv')
    zf = ZipFile(str(x)+'.csv.zip')
    #extracted data is saved in the same directory as notebook
    zf.extractall() 
    zf.close()
    data=pd.read_csv(str(x)+'.csv')
    return data
```

```python
#data handling & cleaning 
df=pd.concat([GetApi(2013),GetApi(2014),GetApi(2015),GetApi(2016),GetApi(2017),GetApi(2018)])
df.drop(['CANCELLED','CANCELLATION_CODE','DIVERTED','CARRIER_DELAY','WEATHER_DELAY','NAS_DELAY','SECURITY_DELAY','LATE_AIRCRAFT_DELAY','Unnamed: 27','AIR_TIME','ACTUAL_ELAPSED_TIME','CRS_ELAPSED_TIME','TAXI_IN','WHEELS_ON','WHEELS_OFF','TAXI_OUT','OP_CARRIER_FL_NUM'], axis=1, inplace=True)
df = df.dropna()
```

```python
df['OP_CARRIER'].replace({
    'UA':'United Airlines',
    'AS':'Alaska Airlines',
    '9E':'Endeavor Air',
    'B6':'JetBlue Airways',
    'EV':'ExpressJet',
    'F9':'Frontier Airlines',
    'G4':'Allegiant Air',
    'HA':'Hawaiian Airlines',
    'MQ':'Envoy Air',
    'NK':'Spirit Airlines',
    'OH':'PSA Airlines',
    'OO':'SkyWest Airlines',
    'VX':'Virgin America',
    'WN':'Southwest Airlines',
    'YV':'Mesa Airline',
    'YX':'Republic Airways',
    'AA':'American Airlines',
    'DL':'Delta Airlines'
},inplace=True)
```

```python
#save dataframe to csv file
df.to_csv("data.csv")
```
